```python
df_krx.groupby(['Market', 'ListingYear'])['Symbol'].count()
```

```python
df_year_market = df_krx.pivot_table(index=['ListingYear', 'Market'], values='Symbol', aggfunc='count')
df_year_market = df_year_market.reset_index()
df_year_market = df_year_market.rename(columns={'Symbol':'count'})
df_year_market
```

```python
df_krx.groupby(['Market', 'ListingYear'])['Symbol'].count().reset_index(name='count')
```

```python
plt.figure(figsize=(15, 4))
sns.scatterplot(data=df_year_market, x='ListingYear', y = 'count', hue='Market', size='count')
```

```python
plt.figure(figsize=(15, 4))
sns.lineplot(data=df_year_market, x = 'ListingYear', y = 'count', ci=None, hue='Market')
```

```python
sns.relplot(data=df_year_market, x='ListingYear', y = 'count', col='Market', hue='Market', kind='line')
```
