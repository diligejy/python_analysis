```python
import pandas as pd
import seaborn as sns
```

```python
bank_df = pd.read_csv('../rawdata/bank_prep.csv')
```

```python
bank_df.head()
```

```python
bank_df.shape
```

```python
bank_df.dtypes
```

```python
# MinMax
from sklearn.preprocessing import MinMaxScaler

bank_df = bank_df.drop('y', axis=1)

mc = MinMaxScaler()
mc.fit(bank_df)

bank_df_mc = pd.DataFrame(mc.transform(bank_df), columns = bank_df.columns)
bank_df_mc.head()
```

```python
# Z Transform
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(bank_df)

bank_df_sc = pd.DataFrame(sc.transform(bank_df), columns=bank_df.columns)
bank_df_sc.head()
```

```python
bank_df_sc.std()
```

```python
# 계층형 클러스터링 그룹분할
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

hcls = linkage(bank_df_sc, metric='euclidean', method='ward')
dendrogram(hcls)

plt.ylabel('dist')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

cst_group = fcluster(hcls, 100, criterion='distance')
print(cst_group)
```

```python
from sklearn.cluster import KMeans

sum_of_sq_dist = {}
for k in range(1, 15):
    km = KMeans(n_clusters=k, init='k-means++', max_iter=1000)
    km = km.fit(bank_df_sc)
    sum_of_sq_dist[k] = km.inertia_

sns.pointplot(x = list(sum_of_sq_dist.keys()), y = list(sum_of_sq_dist.values()))
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Sum of Square Distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans_cls = KMeans(n_clusters=10)
cst_group = kmeans_cls.fit_predict(bank_df_sc)

print(cst_group)
```

```python
for i in range(10):
    labels = bank_df_sc[cst_group == i]
    plt.scatter(labels['age'], labels['balance'], label=i)

plt.legend()
plt.xlabel('age')
plt.ylabel('balance')
plt.show()
```

```python
bank_df_sc['group'] = cst_group
bank_df_sc.head()
```
