```python
import numpy as np
from sklearn.linear_model import LinearRegression as LR
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (20, 5)
```

### 가상 데이터 생성

```python
X = np.linspace(-5, 10, 500) # -5부터 10까지 등간격으로 500개 요소
X_2 = X ** 2 # X_2 : X의 모든 요소에 제곱이 된 형태

Y = 3 * X ** 2 + np.random.normal(0, 4, size = len(X)) # Y는 X^2을 
```

```python
plt.scatter(X, Y)
```

```python
# X는 1차원이어서 sklearn의 인풋 구조와 맞지 않아 reshape사용
# .fit(X, Y) => X = [[record1], [record2], ..., [record n]]으로 해줘야함
# 1차원: X = [record1, record2, ..., record n]
model_1 = LR().fit(X.reshape(-1, 1), Y)
model_2 = LR().fit(X_2.reshape(-1, 1), Y)
```

```python
X.reshape(-1, 1).shape
```

```python
X.shape
```

### 모델 적용

```python
model_1_Y = model_1.predict(X.reshape(-1, 1))
model_2_Y = model_2.predict(X_2.reshape(-1, 1)) # model_2의 입력도 제곱으로 들어감
```

### 결과 시각화

```python
plt.xlabel('X')
plt.ylabel('Y')

plt.plot(X, Y, label = 'data')
plt.plot(X, model_1_Y, label = 'Y = f(x)')
plt.plot(X, model_2_Y, label = 'Y = f(x**2)')

plt.legend(loc = 'upper left')
```
