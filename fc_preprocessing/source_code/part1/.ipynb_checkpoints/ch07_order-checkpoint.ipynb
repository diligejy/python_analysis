```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (30, 10)
```

### 1. 데이터 병합

```python
# 시트명 정의하기
sheet_names = [str(i + 1) + '월' for i in range(12)]
```

```python
merged_df = pd.DataFrame()
for sn in sheet_names:
    df = pd.read_excel(r'C:/Users/Jinyoung/Pictures/python_analysis/fc_preprocessing/rawdata/part1/월별매출데이터.xlsx', sheet_name=sn, skiprows=range(6)).iloc[:, 1:]
    merged_df = pd.concat([merged_df, df], axis=0, ignore_index=True)
```

```python
merged_df.head()
```

### 2. 포맷 통일 및 변수 추가

주문금액 = 수량 * 가격 * 1.1 (부가세)
제품에 따른 가격 정보는 제품별_가격정보.xlsx에 있음

```python
merged_df['일자'] = merged_df['일자'].str.replace('.', '-')
```

```python
merged_df
```

```python
price_df = pd.read_excel(r'C:/Users/Jinyoung/Pictures/python_analysis/fc_preprocessing/rawdata/part1/제품별_가격정보.xlsx')
price_df.head() # 원활한 분석을 위해, 행과 열을 바꿀 필요가 있음
```

```python
price_df = price_df.T
price_df
```

```python
price_df.loc['제품A': '제품F', 0]
```

```python
price_dict = price_df.loc['제품A': '제품F', 0].to_dict()
price_dict
```

```python
# replace : 품명 컬럼에 price_dict의 key가 있으면 value로 바꿈
# 새로운 변수 정의
merged_df['가격'] = merged_df['품명'].replace(price_dict)
```

```python
merged_df.head()
```

```python
# 주문금액 변수 추가
merged_df['주문금액'] = merged_df['수량'] * merged_df['가격'] * 1.1
```

```python
merged_df.head()
```

### 3. 월별 매출 추이 파악

```python
merged_df['월'] = merged_df['일자'].str.split('-', expand=True).iloc[:, 1].astype(int)
```

```python
merged_df.groupby(['월'], as_index=False)['주문금액'].sum()
```

```python
# 월과 품목에 따른 주문 금액 평균 계산
grouped_df_by_month = merged_df.groupby(['월'], as_index=False)['주문금액'].mean()
grouped_df_by_month_and_product = merged_df.groupby(['월', '품명'], as_index=False)['주문금액'].mean()
```

```python
grouped_df_by_month
```

```python
grouped_df_by_month_and_product
```

```python
plt.title('월별 매출 추이')
plt.plot(grouped_df_by_month['월'], grouped_df_by_month['주문금액'], label='총주문금액 평균')

for product in grouped_df_by_month_and_product['품명'].unique():
    plt.plot(grouped_df_by_month_and_product['월'].unique(), # 월의 개수가 12 * 품목개수라는 것에 주의
             grouped_df_by_month_and_product.loc[grouped_df_by_month_and_product['품명'] == product]['주문금액'],
             label=product
            ) 

plt.legend(loc = 'upper right')
```

제품별 판매량 합계를 나타내는 막대 그래프와 제품과 결제 수단에 따른 히트맵

### 5. 충성고객 찾기
주문 금액 합과 빈도가 각각 상위 10% 안에 속하는 고객을 찾아 정리

```python
grouped_df_by_customer = merged_df.groupby('주문인 ID', as_index=False)['주문금액'].agg(['sum', 'count'])
```

```python
threshold_1 = grouped_df_by_customer['sum'].quantile(0.9) # 구매 금액 합계 기준 상위 10%
threshold_2 = grouped_df_by_customer['count'].quantile(0.9) # 구매 빈도 기준 상위 10%

cond1 = grouped_df_by_customer['sum'] >= threshold_1 # 기준 1을 만족하는지 여부
cond2 = grouped_df_by_customer['count'] >= threshold_2 # 기준 2를 만족하는지 여부

grouped_df_by_customer.loc[cond1 & cond2].sort_values(by = ['sum', 'count'], ascending=False)
```
