### 1. 월별 뉴스 기사 수 추이 시각화 

```python
df['월'] = df['작성일자'].str.split('-', expand=True)[1]
```

```python
grouped_df_by_month = df.groupby(['월'], as_index=False)['기사제목'].agg('count')
```

```python
grouped_df_by_month
```

```python
plt.title('월별 뉴스 기사 수')
plt.xlabel('날짜')
plt.ylabel('뉴스 기사 수')
plt.xticks(range(12), [str(i + 1) + '월' for i in range(12)])
plt.ylim(min(grouped_df_by_month['기사제목']) * 0.9, max(grouped_df_by_month['기사제목']) * 1.1)
plt.bar(grouped_df_by_month['월'], grouped_df_by_month['기사제목'])
```

### 주요 단어 추출

```python
word_list = df['기사제목'].str.split(' ', expand = False).sum()
word_list = pd.Series(word_list)
word_list.value_counts().iloc[:10]
```

```python
other_occur = word_list.value_counts().iloc[11:].sum()
major_word_list = word_list.value_counts().iloc[:10]
major_word_list['기타'] = other_occur
```

```python
major_word_list
```

```python
plt.rcParams['figure.figsize'] = (10, 10)
plt.pie(x = [major_word_list['기타'], major_word_list.iloc[:-1].sum()],
       labels = ['기타', '주요단어'],
       colors = ['green', 'blue'],
       labeldistance= 0.8,
       autopct='%1.1f%%')
```

### 3. 월별 주요 단어 출현 빈도 시각화

```python
major_word_list = major_word_list.iloc[:-1] # 기타 제외
```

```python
new_cols = []
for col in major_word_list.index:
    df[col + '_등장횟수'] = df['기사제목'].str.contains(col)
    new_cols.append(col + '_등장횟수')

df.head()
```

```python
grouped_df = df.groupby('월', as_index=False)[new_cols].sum()
```
