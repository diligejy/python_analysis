## DataFrame 정렬하기와 컬럼 바꾸기

```python
df.head()
```

```python
# 인덱스로 정렬하기
df.sort_index().head()
```

```python
# 내림차순 정렬 ascending = False
df.sort_index(ascending = False).head()
```

```python
# sort_values로 정렬하기 df.sort_values(컬럼이름)
# 나이 많은 선수 10명 보여주기
df.sort_values(by='age', ascending = False)[:10].reset_index()
```

```python
# 인덱스를 컬럼이름으로 바꿔보기 df.set_index("컬럼 이름")
# rank를 인덱스로 바꿔보기
df.set_index('rank').head()
```

```python
# 컬럼이름 'team'을 'club'으로 바꾸기
df.rename(columns = {'team' : 'club'}, inplace=True)
```

```python
df.head()
```

## 데이터 전처리

```python
# value값에서 불필요한 문자는 없애고 데이터 타입을 숫자형으로 바꾸기
df['value'] = df['value'].str.replace('€', '')
df['value'] = df['value'].str.replace('m', '')
```

```python
df['value'] = df['value'].astype('float')
```

```python
df.info()
```

```python
df['시장가치(억)'] = df['value'] * 13
df.head()
```

```python
# 컬럼 삭제 df.drop(columns = ['컬럼명'])
df.drop(columns = ['value'], inplace = True)
```

```python
df.head()
```

## DataFrame 통계분석과 Groupby()

```python
# 숫자형 데이터에 대한 통계
df.describe()
```

```python
# 평균 - mean
df['age'].mean()
```

```python
# 몸값 합계 구하기
df['시장가치(억)'].sum()
```

```python
# 선수들이 속한 가장 많은 나라는? 최빈 값 - mode()
df['nation'].mode()
```

```python
df[df['nation'] == 'Brazil']
```

## groupby

```python
df.groupby('nation')
```

```python
df.groupby('nation').size()
```

```python
df.groupby('nation').count()
```

```python
df.groupby('nation').sum()
```

```python
df.groupby('nation')['시장가치(억)'].sum()
```

```python
(df.groupby('nation')['시장가치(억)'].sum() / df.groupby('nation').size()).sort_values(ascending = False)
```

```python
# 선수들의 몸값의 합이 큰 클럽별로 정렬해서 보여주기
df.groupby('club')['시장가치(억)'].sum().sort_values(ascending = False)
```
