## Transfermarkt 크롤링
- 첫페이지 25명 정보

```python
# 필요한 라이브러리 불러오기
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# requests.get()으로 url정보 요청하기
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}           
url = 'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop'
```

```python
r = requests.get(url, headers = headers)
r.status_code
```

```python
# BeautifulSoup으로 웹페이지 분석
# r.content 대ㅐ신 r.text도 가능
soup = BeautifulSoup(r.content, 'html.parser') 
```

```python
# 선수들의 정보가 담긴 태그와 클래스 찾기
player_info = soup.find_all('tr', class_= ['odd', 'even'])
```

```python
# 첫 번째 요소 확인
print(player_info[0].prettify())

# 전체 개수 확인
len(player_info)
```

```python
# 7개 정보를 담을 빈 리스트 만들기 rank, name, position, age, nation, team, value
rank = []
name = []
position = []
age = []
nation = []
team = []
value = []
```

```python
# player_info에서 td태그만 찾기
for info in player_info:
    player = info.find_all('td')
# 해당 정보를 찾아서 각 리스트에 추가
    rank.append(player[0].text)
    name.append(player[1].img['alt'])
    position.append(player[4].text)
    age.append(player[5].text)
    nation.append(player[6].img['alt'])
    team.append(player[7].img['alt'])
    value.append(player[8].b.text)
```

```python
# dataframe으로 변환
df = pd.DataFrame(
        {"rank": rank,
         "name":name,
         "position":position,
         "age":age,
         "nation":nation,
         "team":team,
         "value":value}
    )
df
```
