```python
def extract_year_from_cdiac(value):
    return value.split('-')[0]

df_cda['cdiac_year'] = df_cda['CDIAC Number'].apply(extract_year_from_cdiac)
```

```python
df_cda['cdiac_year']
```

```python
df_cda[df_cda['sale_year'] == 2021]['cdiac_year'].unique()
```

```python
df_cda['cdiac_year'] = df_cda['cdiac_year'].astype(int)
```

```python
df_cda.loc[(df_cda['sale_year'] == 2021) & (df_cda['sale_year'] != df_cda["cdiac_year"])][["sale_year", "cdiac_year"]]
```
